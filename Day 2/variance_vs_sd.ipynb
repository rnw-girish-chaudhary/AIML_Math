```python
# Imports and sample data
import numpy as np
import matplotlib.pyplot as plt

# Two classes: A has low spread, B has high spread
class_A = [70, 72, 68, 65, 90, 85, 88, 92, 60, 75]
class_B = [60, 95, 40, 85, 20, 100, 35, 80, 50, 90]

# Means (averages)
mean_A = np.mean(class_A)
mean_B = np.mean(class_B)

# Sample variance/std (ddof=1 -> divide by n-1)
var_A = np.var(class_A, ddof=1)
var_B = np.var(class_B, ddof=1)
std_A = np.std(class_A, ddof=1)
std_B = np.std(class_B, ddof=1)

print(f"Class A -> mean={mean_A:.1f}, variance={var_A:.1f}, std={std_A:.1f}")
print(f"Class B -> mean={mean_B:.1f}, variance={var_B:.1f}, std={std_B:.1f}")
```

```python
# Variance: Scatter (spread around the mean)
plt.figure(figsize=(6,4))
plt.scatter(range(len(class_A)), class_A, label=f"Class A (Var={var_A:.1f})")
plt.scatter(range(len(class_B)), class_B, label=f"Class B (Var={var_B:.1f})")
plt.axhline(mean_A, linestyle="--", label=f"A mean = {mean_A:.1f}", color="blue")
plt.axhline(mean_B, linestyle="--", label=f"B mean = {mean_B:.1f}", color="orange")
plt.title("Variance via Scatter: distance of points from mean")
plt.xlabel("Student")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
# SD: Scatter showing distance from mean
plt.figure(figsize=(6,4))
plt.scatter(range(len(class_A)), class_A, label=f"Class A (SD={std_A:.1f})")
plt.scatter(range(len(class_B)), class_B, label=f"Class B (SD={std_B:.1f})")
plt.axhline(mean_A, linestyle="--", label=f"A mean = {mean_A:.1f}", color="blue")
plt.axhline(mean_B, linestyle="--", label=f"B mean = {mean_B:.1f}", color="orange")
plt.title("Standard Deviation via Scatter (average distance from mean)")
plt.xlabel("Student")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
# Variance: Histogram (how widely values spread)
plt.figure(figsize=(6,4))
plt.hist(class_A, bins=6, alpha=0.6, label=f"Class A (Var={var_A:.1f})")
plt.hist(class_B, bins=6, alpha=0.6, label=f"Class B (Var={var_B:.1f})")
plt.title("Variance via Histogram: wider spread  higher variance")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# SD: Histogram with mean lines
plt.figure(figsize=(6,4))
plt.hist(class_A, bins=6, alpha=0.6, label=f"Class A (SD={std_A:.1f})")
plt.hist(class_B, bins=6, alpha=0.6, label=f"Class B (SD={std_B:.1f})")
plt.axvline(mean_A, linestyle="--", label=f"A mean = {mean_A:.1f}", color="blue")
plt.axvline(mean_B, linestyle="--", label=f"B mean = {mean_B:.1f}", color="orange")
plt.title("Standard Deviation via Histogram")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Variance: Box plot (compact vs wide box)
plt.figure(figsize=(5,4))
plt.boxplot([class_A, class_B], tick_labels=["Class A", "Class B"])
plt.title("Variance via Box Plot (wider box/whiskers  higher variance)")
plt.ylabel("Score")
plt.show()
```

```python
# SD: Box plot (wider box  higher SD)
plt.figure(figsize=(5,4))
plt.boxplot([class_A, class_B], tick_labels=["Class A", "Class B"])
plt.title("Standard Deviation via Box Plot")
plt.ylabel("Score")
plt.show()
```

```python
# Variance: Line plot (ups/downs across students)
plt.figure(figsize=(6,4))
plt.plot(class_A, marker="o", label=f"Class A (Var={var_A:.1f})")
plt.plot(class_B, marker="o", label=f"Class B (Var={var_B:.1f})")
plt.axhline(mean_A, linestyle="--", label=f"A mean = {mean_A:.1f}", color="blue")
plt.axhline(mean_B, linestyle="--", label=f"B mean = {mean_B:.1f}", color="orange")
plt.title("Variance via Line Plot")
plt.xlabel("Student")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
# SD: Line plot with mean  1 SD shading
# For the band we just need x positions (index numbers)
xA = np.arange(len(class_A))
xB = np.arange(len(class_B))

plt.figure(figsize=(6,4))
plt.plot(class_A, marker="o", label=f"Class A (mean={mean_A:.1f}, SD={std_A:.1f})")
plt.fill_between(xA, mean_A - std_A, mean_A + std_A, alpha=0.2, label="A mean  1 SD")

plt.plot(class_B, marker="o", label=f"Class B (mean={mean_B:.1f}, SD={std_B:.1f})")
plt.fill_between(xB, mean_B - std_B, mean_B + std_B, alpha=0.2, label="B mean  1 SD")

plt.title("Standard Deviation via Line Plot (shaded mean  SD)")
plt.xlabel("Student")
plt.ylabel("Score")
plt.legend()
plt.show()
```
